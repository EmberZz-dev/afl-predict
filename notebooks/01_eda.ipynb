# AFL Match Data — Exploratory Data Analysis

This notebook explores the AFL match dataset (2015–2025) to understand patterns, distributions, and potential predictive signals before building ML models.

**Questions we want to answer:**
1. How many matches per season? Any anomalies?
2. What is the home team win rate?
3. How are scores and margins distributed?
4. Which venues have the strongest home advantage?
5. Which teams are strongest/weakest over time?
6. How do our engineered features correlate with outcomes?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

sns.set_theme(style="darkgrid", palette="deep")
plt.rcParams["figure.figsize"] = (12, 6)
plt.rcParams["figure.dpi"] = 100

PROJECT_ROOT = Path("..").resolve()
```

## 1. Load Data

```python
# Load cleaned match data
matches = pd.read_csv(PROJECT_ROOT / "data" / "processed" / "matches_clean.csv", parse_dates=["date"])
print(f"Dataset: {len(matches)} matches, {matches.columns.size} columns")
print(f"Date range: {matches['date'].min().date()} to {matches['date'].max().date()}")
print(f"Seasons: {sorted(matches['season'].unique())}")
matches.head()
```

```python
matches.describe()
```

## 2. Matches Per Season

We expect ~198 home-and-away games + ~9 finals = ~207 per season. 2020 was COVID-affected (shortened season).

```python
season_counts = matches.groupby("season").size()

fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.bar(season_counts.index, season_counts.values, color="#4a90d9", edgecolor="white")

# Highlight 2020 (COVID)
covid_idx = list(season_counts.index).index(2020)
bars[covid_idx].set_color("#d94a4a")

ax.set_xlabel("Season")
ax.set_ylabel("Number of Matches")
ax.set_title("Matches Per AFL Season (2015–2025)")
ax.axhline(y=207, color="gray", linestyle="--", alpha=0.5, label="Expected (~207)")

for bar, count in zip(bars, season_counts.values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 2, str(count),
            ha="center", va="bottom", fontsize=10, fontweight="bold")

ax.legend()
plt.tight_layout()
plt.show()

print(f"\n2020 had only {season_counts[2020]} matches due to COVID-19 shortened season")
```

## 3. Home Team Win Rate

Home advantage is one of the strongest effects in sport. In AFL, we expect ~57-58% home win rate.

```python
overall_home_wr = matches["is_home_win"].mean()
print(f"Overall home win rate: {overall_home_wr:.1%}")

# Home win rate by season
home_wr_by_season = matches.groupby("season")["is_home_win"].mean()

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(home_wr_by_season.index, home_wr_by_season.values, "o-", color="#5cb85c", linewidth=2, markersize=8)
ax.axhline(y=0.5, color="gray", linestyle="--", alpha=0.5, label="50% (no advantage)")
ax.axhline(y=overall_home_wr, color="#d9a04a", linestyle="--", alpha=0.7, label=f"Overall avg ({overall_home_wr:.1%})")
ax.fill_between(home_wr_by_season.index, 0.5, home_wr_by_season.values,
                 where=home_wr_by_season.values > 0.5, alpha=0.15, color="green")

ax.set_xlabel("Season")
ax.set_ylabel("Home Win Rate")
ax.set_title("AFL Home Team Win Rate by Season")
ax.set_ylim(0.4, 0.7)
ax.legend()

for x, y in zip(home_wr_by_season.index, home_wr_by_season.values):
    ax.annotate(f"{y:.0%}", (x, y), textcoords="offset points", xytext=(0, 10), ha="center", fontsize=9)

plt.tight_layout()
plt.show()
```

## 4. Score & Margin Distributions

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Home scores
axes[0].hist(matches["hscore"], bins=40, color="#4a90d9", edgecolor="white", alpha=0.8)
axes[0].axvline(matches["hscore"].mean(), color="red", linestyle="--", label=f"Mean: {matches['hscore'].mean():.0f}")
axes[0].set_title("Home Team Scores")
axes[0].set_xlabel("Score")
axes[0].legend()

# Away scores
axes[1].hist(matches["ascore"], bins=40, color="#d94a7a", edgecolor="white", alpha=0.8)
axes[1].axvline(matches["ascore"].mean(), color="red", linestyle="--", label=f"Mean: {matches['ascore'].mean():.0f}")
axes[1].set_title("Away Team Scores")
axes[1].set_xlabel("Score")
axes[1].legend()

# Margin distribution
axes[2].hist(matches["margin"], bins=50, color="#5cb85c", edgecolor="white", alpha=0.8)
axes[2].axvline(0, color="black", linestyle="-", linewidth=1.5)
axes[2].axvline(matches["margin"].mean(), color="red", linestyle="--", label=f"Mean: {matches['margin'].mean():.1f}")
axes[2].set_title("Match Margin (Home - Away)")
axes[2].set_xlabel("Margin (pts)")
axes[2].legend()

plt.suptitle("Score & Margin Distributions", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()

print(f"Average home score: {matches['hscore'].mean():.1f}")
print(f"Average away score: {matches['ascore'].mean():.1f}")
print(f"Average margin: {matches['margin'].mean():.1f} (positive = home wins)")
print(f"Median margin: {matches['margin'].median():.1f}")
print(f"Biggest home win: {matches['margin'].max()}")
print(f"Biggest away win: {matches['margin'].min()}")
```

## 5. Venue Analysis — Home Ground Advantage

```python
# Top venues by number of games
venue_stats = matches.groupby("venue").agg(
    games=("margin", "size"),
    home_win_rate=("is_home_win", "mean"),
    avg_margin=("margin", "mean"),
).sort_values("games", ascending=False)

# Only venues with 30+ games
major_venues = venue_stats[venue_stats["games"] >= 30].sort_values("home_win_rate", ascending=True)

fig, ax = plt.subplots(figsize=(10, 8))
colors = ["#d94a4a" if x < 0.5 else "#5cb85c" for x in major_venues["home_win_rate"]]
bars = ax.barh(major_venues.index, major_venues["home_win_rate"], color=colors, edgecolor="white")
ax.axvline(x=0.5, color="gray", linestyle="--", alpha=0.7)
ax.axvline(x=overall_home_wr, color="#d9a04a", linestyle="--", alpha=0.7, label=f"Overall ({overall_home_wr:.0%})")

for bar, rate, games in zip(bars, major_venues["home_win_rate"], major_venues["games"]):
    ax.text(bar.get_width() + 0.005, bar.get_y() + bar.get_height()/2,
            f"{rate:.0%} ({games}g)", va="center", fontsize=9)

ax.set_xlabel("Home Win Rate")
ax.set_title("Home Win Rate by Venue (30+ games)")
ax.legend()
plt.tight_layout()
plt.show()
```

## 6. Team Performance Over Time

```python
# Calculate win rate per team per season
home_wins = matches.groupby(["season", "hteam"]).agg(wins=("is_home_win", "sum"), games=("is_home_win", "size"))
away_wins = matches.copy()
away_wins["is_away_win"] = ~away_wins["is_home_win"].astype(bool)
away_wins = away_wins.groupby(["season", "ateam"]).agg(wins=("is_away_win", "sum"), games=("is_away_win", "size"))

home_wins = home_wins.reset_index().rename(columns={"hteam": "team"})
away_wins = away_wins.reset_index().rename(columns={"ateam": "team"})

all_records = pd.concat([home_wins, away_wins])
team_seasons = all_records.groupby(["season", "team"]).agg(wins=("wins", "sum"), games=("games", "sum")).reset_index()
team_seasons["win_rate"] = team_seasons["wins"] / team_seasons["games"]

# Plot top 6 teams by average win rate
avg_wr = team_seasons.groupby("team")["win_rate"].mean().sort_values(ascending=False)
top_teams = avg_wr.head(6).index.tolist()

fig, ax = plt.subplots(figsize=(12, 6))
for team in top_teams:
    data = team_seasons[team_seasons["team"] == team].sort_values("season")
    ax.plot(data["season"], data["win_rate"], "o-", label=team, linewidth=2, markersize=6)

ax.axhline(y=0.5, color="gray", linestyle="--", alpha=0.5)
ax.set_xlabel("Season")
ax.set_ylabel("Win Rate")
ax.set_title("Top 6 AFL Teams — Win Rate by Season")
ax.legend(bbox_to_anchor=(1.02, 1), loc="upper left")
ax.set_ylim(0.2, 0.9)
plt.tight_layout()
plt.show()
```

```python
# Overall team rankings
fig, ax = plt.subplots(figsize=(10, 8))
avg_wr_sorted = avg_wr.sort_values(ascending=True)
colors = ["#d94a4a" if x < 0.5 else "#4a90d9" for x in avg_wr_sorted.values]
bars = ax.barh(avg_wr_sorted.index, avg_wr_sorted.values, color=colors, edgecolor="white")
ax.axvline(x=0.5, color="gray", linestyle="--", alpha=0.7)

for bar, rate in zip(bars, avg_wr_sorted.values):
    ax.text(bar.get_width() + 0.005, bar.get_y() + bar.get_height()/2,
            f"{rate:.0%}", va="center", fontsize=9, fontweight="bold")

ax.set_xlabel("Average Win Rate (2015–2025)")
ax.set_title("All AFL Teams — Average Win Rate")
plt.tight_layout()
plt.show()
```

## 7. Feature Correlation Analysis

Load the engineered features and check how they correlate with the target (`home_win`).

```python
features = pd.read_csv(PROJECT_ROOT / "data" / "processed" / "features.csv")
print(f"Feature matrix: {features.shape[0]} rows, {features.shape[1]} columns")

# Numeric features only
feature_cols = [
    "home_elo", "away_elo", "elo_diff", "elo_prob",
    "home_form_5", "home_form_10", "away_form_5", "away_form_10",
    "home_avg_margin_5", "away_avg_margin_5",
    "home_avg_score_5", "away_avg_score_5",
    "home_avg_conceded_5", "away_avg_conceded_5",
    "home_streak", "away_streak",
    "h2h_home_wins_5", "h2h_avg_margin_5",
    "home_venue_win_rate", "away_venue_win_rate",
    "round_number", "home_win", "margin",
]
feat_df = features[feature_cols].dropna()
print(f"After dropping NaN: {len(feat_df)} rows")
```

```python
# Correlation with home_win
correlations = feat_df.corr()["home_win"].drop(["home_win", "margin"]).sort_values(key=abs, ascending=False)

fig, ax = plt.subplots(figsize=(10, 8))
colors = ["#5cb85c" if x > 0 else "#d94a4a" for x in correlations.values]
ax.barh(correlations.index, correlations.values, color=colors, edgecolor="white")
ax.axvline(x=0, color="gray", linestyle="-", linewidth=1)

for i, (feat, corr) in enumerate(correlations.items()):
    ax.text(corr + (0.005 if corr > 0 else -0.005), i,
            f"{corr:.3f}", va="center", ha="left" if corr > 0 else "right", fontsize=9)

ax.set_xlabel("Correlation with Home Win")
ax.set_title("Feature Correlation with Match Outcome")
plt.tight_layout()
plt.show()

print("\nTop 5 most predictive features:")
for feat, corr in correlations.head(5).items():
    print(f"  {feat:<25} r = {corr:+.3f}")
```

```python
# Full correlation heatmap
corr_matrix = feat_df[feature_cols[:20]].corr()

fig, ax = plt.subplots(figsize=(14, 11))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt=".2f", cmap="RdBu_r",
            center=0, vmin=-1, vmax=1, square=True, linewidths=0.5,
            ax=ax, annot_kws={"size": 8})
ax.set_title("Feature Correlation Heatmap", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()

# Check for highly correlated features (potential multicollinearity)
print("\nHighly correlated feature pairs (|r| > 0.8):")
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        if abs(corr_matrix.iloc[i, j]) > 0.8:
            print(f"  {corr_matrix.columns[i]} <-> {corr_matrix.columns[j]}: {corr_matrix.iloc[i, j]:.3f}")
```

## 8. ELO Ratings Over Time

```python
# Track ELO evolution for top teams
elo_data = features[["date", "home_team", "away_team", "home_elo", "away_elo"]].dropna()
elo_data["date"] = pd.to_datetime(elo_data["date"])

# Reshape: one row per team per game
home_elo = elo_data[["date", "home_team", "home_elo"]].rename(columns={"home_team": "team", "home_elo": "elo"})
away_elo = elo_data[["date", "away_team", "away_elo"]].rename(columns={"away_team": "team", "away_elo": "elo"})
all_elo = pd.concat([home_elo, away_elo]).sort_values("date")

fig, ax = plt.subplots(figsize=(14, 7))
for team in top_teams:
    team_data = all_elo[all_elo["team"] == team]
    ax.plot(team_data["date"], team_data["elo"], label=team, linewidth=1.5, alpha=0.8)

ax.axhline(y=1500, color="gray", linestyle="--", alpha=0.5, label="Average (1500)")
ax.set_xlabel("Date")
ax.set_ylabel("ELO Rating")
ax.set_title("ELO Rating Evolution — Top 6 Teams")
ax.legend(bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

## 9. ELO as a Predictor

```python
# How well does ELO difference predict outcomes?
elo_outcome = feat_df[["elo_diff", "home_win", "margin"]].copy()
elo_outcome["elo_bin"] = pd.cut(elo_outcome["elo_diff"], bins=10)

binned = elo_outcome.groupby("elo_bin", observed=True).agg(
    home_win_rate=("home_win", "mean"),
    avg_margin=("margin", "mean"),
    count=("home_win", "size")
).reset_index()
binned["elo_mid"] = binned["elo_bin"].apply(lambda x: x.mid)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# ELO diff vs win rate
axes[0].scatter(binned["elo_mid"], binned["home_win_rate"], s=binned["count"]*2, color="#4a90d9", alpha=0.7, edgecolors="white")
axes[0].axhline(y=0.5, color="gray", linestyle="--", alpha=0.5)
axes[0].axvline(x=0, color="gray", linestyle="--", alpha=0.5)
axes[0].set_xlabel("ELO Difference (Home - Away)")
axes[0].set_ylabel("Home Win Rate")
axes[0].set_title("ELO Difference vs Home Win Rate")

# ELO diff vs margin
axes[1].scatter(binned["elo_mid"], binned["avg_margin"], s=binned["count"]*2, color="#5cb85c", alpha=0.7, edgecolors="white")
axes[1].axhline(y=0, color="gray", linestyle="--", alpha=0.5)
axes[1].axvline(x=0, color="gray", linestyle="--", alpha=0.5)
axes[1].set_xlabel("ELO Difference (Home - Away)")
axes[1].set_ylabel("Average Margin")
axes[1].set_title("ELO Difference vs Match Margin")

plt.suptitle("ELO Rating as a Predictor", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()

# Simple ELO accuracy
elo_correct = ((feat_df["elo_diff"] > 0) & (feat_df["home_win"] == 1)) | \
              ((feat_df["elo_diff"] < 0) & (feat_df["home_win"] == 0))
print(f"\nELO-only prediction accuracy: {elo_correct.mean():.1%}")
print(f"(Always predict home wins: {feat_df['home_win'].mean():.1%})")
print(f"ELO beats naive baseline by {elo_correct.mean() - feat_df['home_win'].mean():.1%}")
```

## 10. Key Takeaways

**Summary of findings:**

1. **Home advantage is real** — home teams win ~57-58% of the time across all seasons
2. **2020 was anomalous** — COVID shortened the season and likely affected home advantage
3. **ELO is the strongest single predictor** — ELO difference alone achieves reasonable accuracy
4. **Venue matters** — some grounds have significantly higher home win rates
5. **Form features are correlated with outcomes** — recent performance is predictive
6. **Some features are highly correlated** — ELO-related features share information (expected)

**Implications for modelling:**
- ELO should be the backbone feature
- Form + venue features add incremental value
- May need to handle multicollinearity (tree models handle this naturally)
- 2020 data might hurt or help — worth testing with and without
